```python
from PIL import Image
import numpy as np
from pylab import *
import pca
import imtools
import pickle

path = "/home/tintin/rongheng/datasets/a_thumbs"

img_list = imtools.img_list(path)
```

```python
img = np.array(Image.open(img_list[0])) # open one image to get size
n,m = img.shape[0:2]
num_img = len(img_list)

img_matrix = np.array([np.array(Image.open(img)).flatten() for img in img_list], 'f')
```

```python
# perform PCA
V, S, img_mean = pca.pca(img_matrix)
```

```python
# show some images (mean and 7 first images)
plt.figure()
plt.subplot(2,4,1)
imshow(img_mean.reshape(m,n))
for i in range(7):
    plt.subplot(2,4,i+2)
    imshow(V[i].reshape(m,n))
    
```

```python
# save mean and principal components
with open('font_pca_modes.pkl', 'wb') as f:
    pickle.dump(img_mean, f)
    pickle.dump(V, f)
```

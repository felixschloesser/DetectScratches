# Computing the differnce between two images

```python
from PIL import Image
import matplotlib.pyplot as plt
from PIL.ImageChops import subtract, multiply, screen, difference, add
```

```python

img_1 = Image.open('test_images/bearing_scrached.jpg')
img_2 = Image.open('test_images/bearing_perfect.jpg')

img = difference(img_1, img_2)
img.save("result_images/scrach_difference.jpg")

plt.subplot(131)
plt.imshow(img_1),plt.axis('off')
plt.subplot(132)
plt.imshow(img_2), plt.axis('off')
plt.subplot(133)
plt.imshow(img), plt.axis('off')

```

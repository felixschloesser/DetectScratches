```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import measurements as measure, morphology as morph
import sys
path = "/home/tintin/rongheng/cv/"
sys.path.append(path)
from lib.image_lib import *
from lib.edge_lib import threshold
```

```python
img = load_img('houses.png')
show_img(img)
print(img.shape)
show_hist(img)
```

```python
thresh_img = threshold(img, 200)
binary_img = np.logical_not(thresh_img.astype(bool))
show_img(binary_img)
show_hist(binary_img.astype(int))
```

```python
labels, num_obj = measure.label(binary_img)
print("Number of labels: ", num_obj)
```

```python
show_img(labels)
```

```python
# morphology - opening to seperate better
struc_elem = np.ones((12,5))
show_img(struc_elem, size=((1,1)))
img_open = morph.binary_opening(binary_img, struc_elem, iterations=2)
print(img_open.shape, img_open)
labels_open, num_obj_open = measure.label(img_open)
print("Number of objects:", num_obj_open)
```

```python
show_img(labels_open, size=(10,10))
```

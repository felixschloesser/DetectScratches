```python
import numpy as np
import matplotlib.pyplot as plt

sample_frequency = 1000
time = np.linspace(0,2*np.pi, sample_frequency)

def plot_signal(sig, amp, title=""):
    global time
    plt.figure(figsize=(10,2))
    plt.plot(time,sig)
    plt.ylim((-amp,+amp))
    plt.xlabel("Time ($s$)")
    plt.ylabel("Aplitude ($Unit$)")
    if title:
        plt.title(title)
```

```python
freq = 5
amp = 3

sig_1 = amp * np.sin(2*np.pi * freq * time) 
plot_signal(sig_1, 3)
```

```python
freq = 11
amp = 1

sig_2 = amp * np.sin(2*np.pi * freq * time)
plot_signal(sig_2, 3)
```

```python
signal = sig_1 * sig_2
plot_signal(signal, 4)
```

```python
noise_level = 3
noise = noise_level * np.random.ranf(len(time)) - 1.5
plot_signal(noise,4)

signal_noisy = signal + noise
plot_signal(signal_noisy,4)
```

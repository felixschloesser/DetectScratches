```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import filters
import sys
sys.path.append("/home/tintin/rongheng/cv/")
from imtools import load_img, show_img
from edge_detect_lib import harris_response, get_points, overlay_points, threshold
```

```python
plt.style.use('dark_background')
```

```python
img = load_img("oriental_pearl.jpg")
```

```python
response = harris_response(img, sigma=2)
show_img(response, color=True)

thresh = threshold(response, 0.3, percent=True)
points = get_points(thresh, min_dist = 20)
overlay_points(img, points, size=(20,20))
```

## Upsampling

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np 
```

```python
file = 'test_images/empire.jpg'
img = Image.open(file)
plt.figure(figsize=(20,20))
plt.imshow(img)
plt.axis('off')
```

```python
img.thumbnail((128,128))
#img = np.array(img)
plt.imshow(img)
plt.axis('off')
```

### Nearest Neighbor

```python
img_1 = img.resize((img.width*5, img.height*5), Image.NEAREST)
plt.figure(figsize=(10,10))
plt.imshow(img_1)
```

### Bilinear

```python
img_1 = img.resize((img.width*5, img.height*5), Image.BILINEAR)
plt.figure(figsize=(10,10))
plt.imshow(img_1)
```

## Bicubic

```python
img_1 = img.resize((img.width*5, img.height*5), Image.BILINEAR)
plt.figure(figsize=(10,10))
plt.imshow(img_1)
```

```python
methods = ['none', 'nearest', 'bilinear', 'bicubic', 'spline16', 'lanczos']
fig, axes = plt.subplots(2, 3, figsize=(15,10), subplot_kw={'xticks':[], 'yticks':[]})
fig.subplots_adjust(hspace=0.05, wspace=0.05)
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(img, interpolation=interp_method)
    ax.set_title(str(interp_method), size=20)
plt.tight_layout()
plt.show()
```

## Downsampling

```python
img = Image.open('test_images/empire.jpg')
img_down = img.resize((img.width//5, img.height//5))
plt.figure(figsize=(15,10))
plt.imshow(img_down)
```

```python
img_antialias = img.resize(((img.width//5), (img.height//5)), resample=Image.LANCZOS)
plt.figure(figsize=(15,10))
plt.imshow(img_antialias)
```

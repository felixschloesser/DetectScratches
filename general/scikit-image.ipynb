```python
import os
from skimage import *
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
```

```python
def show_img(img):
    plt.figure(figsize=(5,5))
    plt.axis('off')
    plt.imshow(img)
    
def show_bw_img(img):
    plt.figure(figsize=(5,5))
    plt.axis('off')
    plt.imshow(img, cmap='gray', vmin=0, vmax=1)
```

```python
file = 'oriental_pearl.jpg'
img = plt.imread(file) # read image into array
print(img.shape, img.dtype, type(img))
img_hsv = color.rgb2hsv(img)
img_hsv[:,:,1] = 0.5 # change the saturation
img_desat = color.hsv2rgb(img_hsv) # back to rgb

show_img(img_desat)
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import imtools
```

```python
filename = 'bearing_soft.jpg'

img = Image.open(filename).convert('L')
img.thumbnail((512,512))
img_array = np.array(img, dtype='f')
```

## Histogramm Normaisation

```python
plt.hist(img_array.flatten(), bins = 256)
plt.show()
plt.imshow(img_array, cmap='gray', vmin=0, vmax=255)
```

```python
def eq_hist(img, num_bins = 256):
    shades_of_gray = num_bins
    """
    Histogramm equilisation of grayscale image
    """
    hist, bins, _ = plt.hist(img.flatten(), num_bins, density=True);
    cdf = hist.cumsum();
    cdf = cdf / cdf[-1] * shades_of_gray; # normalize
    
    img_2 = np.interp(img.flatten(), bins[:-1], cdf);
    plt.cla()

    return img_2.reshape(img.shape)
```

```python
fig, ax = plt.subplots()

ax.hist(eq_hist(img_array).flatten(), bins=256);
ax.set_xlim([0,255]);
ax1 = ax.twinx();
ax1.hist(eq_hist(img_array).flatten(), bins=256, cumulative=True, density=True, histtype='step', color='tab:red')
```

```python
fig, ax = plt.subplots(1,2, constrained_layout = True)
ax[0].imshow(img_array, cmap='gray', vmin=0, vmax=255)
ax[0].axis('off')
ax[1].imshow(eq_hist(img_array), cmap='gray', vmin=0, vmax=255)
ax[1].axis('off')
```

### Averaging Images

```python
def compute_average(img_list):
    """
    Compute the average of a list of images
    """
    # open the first image and make it into an array of floats
    
    average_img = array(Image.open(img_list[0]),'f')
    
    for img in img_list[1:]:
        try:
            average_img += array(Image.open(img))
        except:
                print(img + '...skipped')
    
    average_img /= len(img_list)
    
    # return the average as an uint8
    return array(average_img, 'uint8')
```

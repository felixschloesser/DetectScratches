```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import imtools
```

```python
def show_img(img):
    plt.figure(figsize=(5,5))
    plt.axis('off')
    plt.imshow(img)
```

```python
file = 'test_images/parrot.jpg'
# read image into array
img = np.array(Image.open(file))
show_img(img)
```

```python
plt.imshow(img)

# some points
x = [100, 100, 300, 300]
y = [100, 330, 200, 330]

# plot the points with red star-markers
plt.plot(x, y, 'y*')
plt.plot(x[:2], y[:2], 'r')

plt.axis('off')

plt.title("Plotting: '" + file +"'")
```

```python
plt.figure()
plt.hist(img.flatten(), 128)
plt.show()
```

```python
img = Image.open(file)
img.thumbnail((512,512))
img = np.array(img)
ch_red = img[:,:,0]
ch_green = img[:,:,1]
ch_blue = img[:,:,2]
```

```python
fig, ax = plt.subplots()
ax.hist(ch_red.flatten(), bins=256, color='r', alpha=0.5)
ax.hist(ch_green.flatten(), bins=256, color='g', alpha=0.4)
ax.hist(ch_blue.flatten(), bins=256, color='b', alpha=0.3)
plt.show()
```

```python
fig, p = plt.subplots(1,3)

p[0].imshow(ch_red, cmap=plt.cm.Reds)
p[0].axis('off')
p[1].imshow(ch_green, cmap=plt.cm.Greens)
p[1].axis('off')
p[2].imshow(ch_blue, cmap=plt.cm.Blues)
p[2].axis('off')

plt.show()
```

```python
img_ch_red = Image.fromarray(ch_red, 'L')
img_ch_green = Image.fromarray(ch_green, 'L')
img_ch_blue = Image.fromarray(ch_blue, 'L')


img = Image.merge('RGB', (img_ch_blue, img_ch_red ,img_ch_green))
img.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import filters
```

```python
with Image.open('test_images/oriental_pearl.jpg') as I:
    img_bw = np.array(I.convert('L'))
    img_color = np.array(I)
```

```python
sigmas = [0,2,5,10]
frames = tuple('frame' + str(i) for i in range(len(sigmas)) )

fig, frames = plt.subplots(1,len(frames), figsize=(20,20))
for frame in frames:
    sigma = sigmas.pop(0)
    img_blurred = filters.gaussian_filter(img_bw, sigma)
    frame.imshow(img_blurred, cmap='gray', vmin=0, vmax=255)
    frame.axis('off')
    frame.set_title(' sigma = ' + str(sigma))
```

```python
img_2 = np.zeros(img_color.shape)
for i in range(3):
    img_2[:,:,i] = filters.gaussian_filter(img_color[:,:,i],5)
img_2 = np.uint8(img_2)
```

```python
plt.imshow(img_2), plt.axis('off')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import measurements as measure, morphology as morph
import sys
path = "/home/tintin/rongheng/cv/"
sys.path.append(path)
from imtools import *
from edge_detect_lib import threshold
```

```python
img = load_img('houses.png')
show_img(img, color=True)
print(img.shape)
show_hist(img)
```

```python
img = threshold(img, 240, invert=True)
show_img(img, color=True)
show_hist(img)
```

```python
labels, num_obj = measure.label(img)
print("Number of labels: ", num_obj)
```

```python
show_img(labels, color=True)
```

```python
# morphology - opening to seperate better
struc_elem = np.ones((12,5))
show_img(struc_elem, size=((1,1)))
img_open = morph.binary_opening(img, struc_elem, iterations=2)

labels_open, num_obj_open = measure.label(img_open)
print("Number of objects:", num_obj_open)
```

```python
show_img(labels_open, size=(10,10), color=True)
```
